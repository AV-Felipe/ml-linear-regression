# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('./Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #here we are creating the regressor, that will be used to generate our model - regression is dieferent fron classification, operates on a continuous set, classification on grouping of things
regressor.fit(x_train, y_train) #the fit method is used to connect the model to the dataset and performing the training of the model, we are passing it 2 two-dimensional arrays, x with the features dataset, an y with the dependnet variables dataset - this arrays are 2d because each vector, within the array, represents a row, which may contain many columns
```

## Predicting the Test set results

```python
y_pred = regressor.predict(x_test) # here we are creating the predictions for the dependent variables for the test set features
```

## Visualising the Training set results

```python
plt.scatter(x_train, y_train, color = 'red') #here we are ploting the values used for the training set as dots (scattered) in a 2d plot
plt.plot(x_train, regressor.predict(x_train), color = 'blue') #here we are ploting a line of prediction based on the training series, that should be the optimum line
plt.title('Salary x Experience (training set)') #we are adding an title to the graphic
plt.xlabel('years of experience') #labeling the x vector
plt.ylabel('salary') #labeling the y vector
plt.show() #here we are actually ploting the graph
```

## Visualising the Test set results

```python
plt.scatter(x_test, y_test, color = 'red') #here we are ploting the values used for the test set as dots (scattered) in a 2d plot, we are using the actual values, here
plt.plot(x_train, regressor.predict(x_train), color = 'blue') #in a linar prediction model we don´t have to change the prediction lines, because, acctually, they will always be equal to the training pattern (we can to see, if we want)
plt.title('Salary x Experience (test set)') #we are adding an title to the graphic
plt.xlabel('years of experience') #labeling the x vector
plt.ylabel('salary') #labeling the y vector
plt.show() #here we are actually ploting the graph
```

```python
plt.scatter(x_test, y_test, color = 'red') #here we are ploting the values used for the test set as dots (scattered) in a 2d plot, we are using the actual values, here
plt.plot(x_test, y_pred, color = 'blue') #in a linar prediction model we don´t have to change the prediction lines, because, acctually, they will always be equal to the training pattern (we can to see, if we want)
plt.title('Salary x Experience (test set)') #we are adding an title to the graphic
plt.xlabel('years of experience') #labeling the x vector
plt.ylabel('salary') #labeling the y vector
plt.show() #here we are actually ploting the graph
```

## Making a single prediction

Our standard feature is composed by one column only, each row containing a number of worked years, so, to predict the salary for a particular
number of worked years, we just ned to pass the desired date to the predict method (in the 2d arrray format)

```python
employement_time = [[12]] # we are defining a feature dataset on which we want our model to operate a prediction, our feature data is composed by one colum and our desired amostration is compose of only one row

predicted_Salary = regressor.predict(employement_time) #we are asking to our model to perform an prediction based on its former training

print(f'the expected wage for this employe is of ${predicted_Salary[0]:.2f}') # we are printing the prediction aftrer extracting it from its array and setting ist format to 2 decimal places precision (:.2f)
```

## Geting the final linear regression equation for our model

Our model consist of an linear equation relating some data, so we can obtain this equation, based on our dataset

![image](./assets/slope-intercept-linear-equation.png)

```python
slope = regressor.coef_
intercept = regressor.intercept_

print(f'y = {slope[0]} . x + {intercept}')
```
